```python
import pandas as pd
import numpy as np
import subprocess
import random
from scipy import stats
import glob
import math
import csv
import sys
import os

import matplotlib.pyplot as plt
from matplotlib import animation
import seaborn as sns
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams['font.family'] = "serif"
%matplotlib inline
```

```python
# declarations
base_dir_data = "/ccb/salz8-1/avaraby/tx_noise/data/"
base_dir_out = "/ccb/salz8-1/avaraby/tx_noise/full_analysis_t3_s10_26022020/GTEx_aggs/"
out_dir = "/ccb/salz8-1/avaraby/tx_noise/full_analysis_t3_s10_26022020/sim_samples/"
hg38_fa = "/home/avaraby1/genomes/human/hg38/hg38_p8.fa"

num_tissues = 3
num_samples = 10

readlen = 101

gffread_path = "gffread"
genRNAseq_path = "/ccb/salz8-1/avaraby/tx_noise/genRNAseq_noFold_cov2.R"
shuffleReads_path = "/ccb/salz8-1/avaraby/tx_noise/shuffleReads.sh"

gff3cols=["seqid","source","type","start","end","score","strand","phase","attributes"]
```

```python
# first extract fasta sequences for each annotation
for tissue_num in range(num_tissues):
    print("\n=================\nTissue #"+str(tissue_num)+"\n=================\n")
    for sample_num in range(num_samples):
        print("++++++\n>Sample #"+str(sample_num)+"\n++++++\n")
        cmd_real = [gffread_path,
                    "-w",out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    "-g",hg38_fa,
                    out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"]
        subprocess.call(cmd_real)

        cmd_nonint = [gffread_path,
                    "-w",out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    "-g",hg38_fa,
                    out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"]
        subprocess.call(cmd_nonint)

        cmd_int = [gffread_path,
                    "-w",out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    "-g",hg38_fa,
                    out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"]
        subprocess.call(cmd_int)

        cmd_pol = [gffread_path,
                    "-w",out_dir+"intergenic.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    "-g",hg38_fa,
                    out_dir+"intergenic.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"]
        subprocess.call(cmd_pol)
```

```python
# simulate reads
for tissue_num in range(num_tissues):
    print("\n=================\nTissue #"+str(tissue_num)+"\n=================\n")
    for sample_num in range(num_samples):
        print("++++++\n>Sample #"+str(sample_num)+"\n++++++\n")
        if not os.path.exists(out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)):
            os.makedirs(out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num))
        cmd_real = [genRNAseq_path,
                    out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+".cov",
                    str(readlen),
                    out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+"/",
                    "0"]
        subprocess.call(cmd_real)

        if not os.path.exists(out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)):
            os.makedirs(out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num))
        cmd_nonint = [genRNAseq_path,
                    out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)+".cov",
                    str(readlen),
                    out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)+"/",
                    "0"]
        subprocess.call(cmd_nonint)

        if not os.path.exists(out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)):
            os.makedirs(out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num))
        cmd_int = [genRNAseq_path,
                    out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)+".cov",
                    str(readlen),
                    out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)+"/",
                    "0"]
        subprocess.call(cmd_int)

        if not os.path.exists(out_dir+"intergenic.t"+str(tissue_num)+"_s"+str(sample_num)):
            os.makedirs(out_dir+"intergenic.t"+str(tissue_num)+"_s"+str(sample_num))
        cmd_pol = [genRNAseq_path,
                    out_dir+"intergenic.t"+str(tissue_num)+"_s"+str(sample_num)+".fasta",
                    out_dir+"intergenic.t"+str(tissue_num)+"_s"+str(sample_num)+".cov",
                    str(readlen),
                    out_dir+"intergenic.t"+str(tissue_num)+"_s"+str(sample_num)+"/",
                    "0"]
        subprocess.call(cmd_pol)
```

```python
# combine annotations for processing with sim2sam
for tissue_num in range(num_tissues):
    print("\n=================\nTissue #"+str(tissue_num)+"\n=================\n")
    for sample_num in range(num_samples):
        print("++++++\n>Sample #"+str(sample_num)+"\n++++++\n")
        merge_cmd = "cat "+out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"+\
                           out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"+\
                           " > "+out_dir+"real_splicing.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"
        subprocess.call(merge_cmd,shell=True)
        
        merge_cmd = "cat "+out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"+\
                           out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"+\
                           " > "+out_dir+"real_intronic.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"
        subprocess.call(merge_cmd,shell=True)
        
        merge_cmd = "cat "+out_dir+"real.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"+\
                           out_dir+"splicing.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"+\
                           out_dir+"intronic.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"+\
                           " > "+out_dir+"all.t"+str(tissue_num)+"_s"+str(sample_num)+".gtf"
        subprocess.call(merge_cmd,shell=True)        
```
